Great! Have regigged this to work with Python3.6. And I can use `tickers.keys()` to extract all of the currency pairs and pass these onto getting candle chart data. However, note that the PoloniexAPI.py file does not have returnChartData() implemented, I will need to do this myself.

With returnChartData working, let's get the chart data for all currency pairs.

```python
from collections import defaultdict

chart_data = defaultdict()

for i, x in enumerate(tickers.keys()):
    if i < 10:
        print("Getting {} of {} currency pairs".format(i, len(tickers.keys())))
        chart_data[x] = api.returnChartData(x, '2017-01-01 00:00:00', '2017-04-14 11:30:00', 900)
```
