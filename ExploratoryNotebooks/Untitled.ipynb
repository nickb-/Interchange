```python
import os

os.chdir("../CodeBase")

from PoloniexAPI import *
import json

with open("../Secrets/secrets.json") as f:
    secrets = json.load(f)
    
api = poloniex(secrets['poloniex']['key'], secrets['poloniex']['secret'])

from Downloader import *
dw = Downloader('poloniex', data_directory = '../Data/2016')
```

```python
from collections import defaultdict

data_file_pattern = "../Data/{}/poloniex/PoloniexChartData_USDT_BTC.csv"
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

for yr in years:
    data = api.returnChartData("USDT_BTC", '{}-01-01 00:00:00'.format(yr), '{}-12-31 11:59:59'.format(yr), 900)
    data = pd.io.json.json_normalize(data)
    data['Datestamp'] = [datetime.datetime.utcfromtimestamp(x).strftime(format) for x in data['date']]
    
    print("-------------")
    print(yr)
    print(data.head())
    print(data.tail())
    
    filename = data_file_pattern.format(yr)
    print("Writing out to: {}".format(filename))
    print("")
    data.to_csv(filename, index = False)
```
